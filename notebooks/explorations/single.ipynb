<br>

> * A one-day-ahead prediction

<br>

Hence, the window arguments

```python
arguments = Arguments(input_width=1, label_width=1, shift=1, 
                      training_=training_, validating_=validating_, testing_=testing_, 
                      label_columns=['estimatedNewAdmissions'])
```

## Modelling

```python
column_indices = {name: i for i, name in enumerate(reduced.columns)}
logger.info(column_indices)
```

<br>

Initiate a generic modelling steps instance

```python
steps = src.prototyping.ModellingSteps.ModellingSteps()
```

<br>

An extension of WindowGenerator for graph drawing

```python
graph_extended = src.prototyping.WindowGraphs.WindowGraphs(input_width=ae.input_width, label_width=ae.label_width, shift=ae.shift, 
                                                    training=ae.training_, validating=ae.validating_, testing=ae.testing_, 
                                                    label_columns=ae.label_columns)
```

```python
graph_convolved = src.prototyping.WindowGraphs.WindowGraphs(input_width=ac.input_width, label_width=ac.label_width, shift=ac.shift, 
                                                     training=ac.training_, validating=ac.validating_, testing=ac.testing_, 
                                                     label_columns=ac.label_columns)
```

<br>

### Baseline

```python
baseline = src.prototyping.ModellingBaseline.ModellingBaseline(label_index=column_indices['estimatedNewAdmissions'])
```

```python
baseline.compile(loss=tf.losses.MeanSquaredError(), metrics=[tf.metrics.MeanAbsoluteError()])

validation_performance = {}
performance = {}

validation_performance['Baseline'] = baseline.evaluate(window_simple.validate)
performance['Baseline'] = baseline.evaluate(window_simple.test, verbose=0)
```

```python
graph_extended.plot(baseline)
```

<br>

### Linear

```python
linear = tf.keras.Sequential([tf.keras.layers.Dense(units=1)])
```

```python
history = steps.modelling(model=linear, window=window_simple)

validation_performance['Linear'] = linear.evaluate(window_simple.validate)
performance['Linear'] = linear.evaluate(window_simple.test, verbose=0)
```

```python
graph_extended.plot(linear)
```

<br>
<br>

### Dense

```python
dense = tf.keras.Sequential([tf.keras.layers.Dense(units=64, activation='relu'), tf.keras.layers.Dense(units=64, activation='relu'), 
                             tf.keras.layers.Dense(units=1)])

history = steps.modelling(model=dense, window=window_simple)

validation_performance['Dense'] = dense.evaluate(window_simple.validate)
performance['Dense'] = dense.evaluate(window_simple.test, verbose=0)
```

```python
graph_extended.plot(dense)
```

<br>
<br>

### Convolution

```python
convolution = tf.keras.Sequential([
    tf.keras.layers.Conv1D(filters=32, kernel_size=(convolution_width,), activation='relu'), 
    tf.keras.layers.Dense(units=32, activation='relu'), 
    tf.keras.layers.Dense(units=1)])

history = steps.modelling(model=convolution, window=window_convolved)

validation_performance['Convolution'] = convolution.evaluate(window_convolved.validate)
performance['Convolution'] = convolution.evaluate(window_convolved.test, verbose=0)
```

```python
graph_convolved.plot(convolution)
```

<br>
<br>

### RNN

Note:

* Shape [batch, time, features] $\Rightarrow$ [batch, time, lstm units]
* Because `return_sequences=True`, the model is fed 18 days of training data at a time

```python
lstm = tf.keras.models.Sequential(
    [tf.keras.layers.LSTM(32, return_sequences=True), tf.keras.layers.Dense(units=1)])

history = steps.modelling(model=lstm, window=window_extended)

validation_performance['LSTM'] = lstm.evaluate(window_extended.validate)
performance['LSTM'] = lstm.evaluate(window_extended.test, verbose=0)
```

```python
graph_extended.plot(lstm)
```

```python
logger.info('Input Shape: %s', window_extended.example[0].shape)
logger.info('Output Shape: %s', lstm(window_extended.example[0]).shape)
```
